```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.dates as mdates
import datetime
```

```python
colors = ['#006FED', '#E03424', 'gray', '#009966', '#000866', '#336600', '#006633', 'm', 'r']
```

```python
total_occupancy = pd.read_csv('covid_hospital_activity/total_beds.csv')
total_occupancy['Date'] = pd.to_datetime(total_occupancy['Date'])
```

```python
total_occupancy
```

```python
plt.figure(figsize=(18,10))
for i, region in enumerate(total_occupancy.columns.values[2:]):
    plt.plot(total_occupancy['Date'], total_occupancy[region], label=region, color=colors[i])
plt.title('COVID Total Bed Occupancy')
plt.legend()
plt.xlabel('Date')
ax = plt.gca()
ax.set_xlim([datetime.datetime(2020, 12, 1), datetime.datetime(2020, 12, 31)])
ax.xaxis.set_major_locator(mdates.DayLocator(interval=1))
ax.xaxis.set_major_formatter(mdates.DateFormatter('%d %b'))
plt.gcf().autofmt_xdate() # Rotation
plt.grid()
plt.savefig('plots/total_occupancy')
```
